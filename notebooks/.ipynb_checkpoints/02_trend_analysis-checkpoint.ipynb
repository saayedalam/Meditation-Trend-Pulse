# üìò Notebook 2: Data Cleaning & Preparation

In this notebook, we load and clean the three main datasets used in this project:

1. **Google Trends - Weekly Interest Over Time**
   - Cleaned column names and date index
   - Applied rolling average smoothing
   - Detected top spikes in search interest
   - Calculated % change from start to end
   - Prepared seasonal heatmap structure

2. **Google Trends - Country-Level Interest**
   - Cleaned columns and renamed for clarity
   - Pivoted to create a keyword √ó country matrix
   - Saved both long and wide versions

3. **Google Trends - Related Queries**
   - Cleaned column names and standard values
   - Split into "top" vs "rising" types
   - Pivoted for keyword √ó query_type √ó score
   - Saved both tidy and pivoted versions

The cleaned data is saved to the `/data/processed/` folder for use in Tableau or downstream analysis.

```python
# üì¶ Imports
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# üìä Load saved datasets
trend_path = "../data/raw/interest_over_time.csv"
country_path = "../data/raw/interest_by_country.csv"
related_path = "../data/raw/related_queries.csv"

df_trends = pd.read_csv(trend_path, parse_dates=True, index_col=0)
df_country = pd.read_csv(country_path)
df_related = pd.read_csv(related_path)
```

```python
# üîç Quick inspection
print("üïí df_trends:", df_trends.shape)
print(df_trends.head(), "\n")

print("üåç df_country:", df_country.shape)
print(df_country.head(), "\n")

print("üîó df_related:", df_related.shape)
print(df_related.head())
```

```python
# üîç Inspect structure
df_trends.info()
df_trends.head()

# ‚úÖ Ensure index is datetime
df_trends.index = pd.to_datetime(df_trends.index)

# ‚úÖ Check for missing values
missing = df_trends.isnull().sum()
print("Missing values per column:\n", missing)

# ‚úÖ Check for duplicates
duplicates = df_trends.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
```

```python
# üìä Apply Seaborn styling for cleaner, more aesthetic plots
sns.set(style="whitegrid", palette="Set2", font_scale=1.1)

# Plot: Raw Weekly Search Interest
df_trends.plot(figsize=(14, 6), linewidth=2)
plt.title("Weekly Search Interest: Meditation & Related Keywords", fontsize=16, weight='bold')
plt.ylabel("Search Interest (0-100)")
plt.xlabel("Date")
plt.legend(title="Keyword", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Plot: Smoothed 4-Week Rolling Average
df_trends.rolling(4).mean().plot(figsize=(14, 6), linewidth=2)
plt.title("Smoothed Weekly Search Interest (4-Week Avg)", fontsize=16, weight='bold')
plt.ylabel("Search Interest (Smoothed)")
plt.xlabel("Date")
plt.legend(title="Keyword", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# ‚úÖ Identify top 3 peak dates per keyword (long-form output)

# 1. Melt wide to long
df_trends_long = df_trends.reset_index().melt(id_vars="date", var_name="keyword", value_name="search_interest")

# 2. Sort and get top 3 per keyword
df_top_peaks = (
    df_trends_long
    .sort_values(["keyword", "search_interest"], ascending=[True, False])
    .groupby("keyword")
    .head(3)
    .reset_index(drop=True)
)

# ‚úÖ Save cleaned version
df_top_peaks.to_csv("../data/processed/interest_top_peaks.csv", index=False)
```

```python
# üìà Total percent change in interest from start to end of period
df_pct_change = ((df_trends.iloc[-1] - df_trends.iloc[0]) / df_trends.iloc[0]) * 100
df_pct_change = df_pct_change.round(2)
df_pct_change
```

```python
# üìâ Smoothed weekly trends using 4-week rolling average
df_trends_smoothed = df_trends.rolling(window=4).mean()
df_trends_smoothed.head()
```

```python
# üìä Monthly average search interest for heatmap
df_monthly_avg = df_trends.copy()
df_monthly_avg["year"] = df_monthly_avg.index.year
df_monthly_avg["month"] = df_monthly_avg.index.month

df_heatmap = df_monthly_avg.groupby(["year", "month"]).mean().unstack()
df_heatmap
```

```python
# Purpose: Save cleaned trend analysis outputs to processed data folder
os.makedirs("../data/processed", exist_ok=True)

# Rolling 4-week average (smoothed trend)
df_trends_smoothed.to_csv("../data/processed/interest_over_time_smoothed.csv")

# Percentage change from first to last point
df_pct_change.to_csv("../data/processed/interest_percent_change.csv", header=['pct_change'])

# Monthly average for heatmap
df_heatmap.to_csv("../data/processed/interest_monthly_heatmap.csv")
```

```python
# üëÅÔ∏è Inspect basic structure
df_country.info()

# üîç Preview the first few rows
df_country.head()

# üßÆ Count unique keywords
df_country['keyword'].value_counts()
```

```python
# üöÆ Drop 'Unnamed: 0' column
df_country = df_country.drop(columns='Unnamed: 0')

# üßπ Drop duplicates if any
df_country = df_country.drop_duplicates()

# üßº Rename columns for clarity (optional but good practice)
df_country = df_country.rename(columns={
    'interest': 'search_interest'
})

# üß≠ Reorder columns (optional)
df_country = df_country[['country', 'keyword', 'search_interest']]

# ‚úÖ Confirm shape and preview
print("‚úÖ Cleaned Shape:", df_country.shape)
df_country.head()
```

```python
# üåç Pivot table: Rows = country, Columns = keyword, Values = interest
df_country_pivot = df_country.pivot_table(
    index='country',
    columns='keyword',
    values='search_interest',
    fill_value=0
).reset_index()

# ‚úÖ Preview
print("‚úÖ Pivoted Shape:", df_country_pivot.shape)
df_country_pivot.head()
```

```python
# üíæ Save the cleaned long-form dataset
df_country.to_csv("../data/processed/country_interest_long.csv", index=False)

# üíæ Save the pivoted dataset for mapping/visualization
df_country_pivot.to_csv("../data/processed/country_interest_pivot.csv", index=False)
```

```python
df_related.info()
df_related.head()
df_related['keyword'].value_counts()
df_related['type'].value_counts()
```

```python
# Drop index column and rename for consistency
df_related_cleaned = df_related.drop(columns='Unnamed: 0').rename(columns={
    'keyword': 'keyword',
    'query': 'related_query',
    'type': 'query_type',
    'value': 'popularity_score'
})

# Preview
df_related_cleaned.head()
```

```python
# Optional pivot to see top queries per keyword and type
df_pivot_related = df_related_cleaned.pivot_table(
    index='related_query',
    columns=['keyword', 'query_type'],
    values='popularity_score',
    fill_value=0
)

# Preview
df_pivot_related.head()
```

```python
# Save cleaned related queries
df_related_cleaned.to_csv("../data/processed/related_queries_cleaned.csv", index=False)

# Optional: Save pivoted version for inspection or export
df_pivot_related.to_csv("../data/processed/related_queries_pivot.csv")
```
